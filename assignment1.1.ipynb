1.4. Create a function in both TensorFlow and PyTorch that takes two tensors as input and returns the sum of their elements. Test the function with 1D, 2D, and 3D tensors. This function will demonstrate how to perform element-wise addition between two tensors and then sum all elements of the resulting tensor.

````python
# TensorFlow test tensors

# one-dimensional tensors
tensor1_tf = tf.constant([1, 2, 3])
tensor2_tf = tf.constant([4, 5, 6])

# two-dimensional tensors
tensor1_tf_2d = tf.constant([[1, 2], [3, 4]])
tensor2_tf_2d = tf.constant([[5, 6], [7, 8]])

tensor1_tf_3d = tf.constant([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
tensor2_tf_3d = tf.constant([[[9, 10], [11, 12]], [[13, 14], [15, 16]]])

# PyTorch test tensors

# one-dimensional tensors
tensor1_pt = torch.tensor([1, 2, 3])
tensor2_pt = torch.tensor([4, 5, 6])

# two-dimensional tensors
tensor1_pt_2d = torch.tensor([[1, 2], [3, 4]])
tensor2_pt_2d = torch.tensor([[5, 6], [7, 8]])

# three-dimensional tensors
tensor1_pt_3d = torch.tensor([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
tensor2_pt_3d = torch.tensor([[[9, 10], [11, 12]], [[13, 14], [15, 16]]])


# Define the function in TensorFlow
def tensor_sum_tf(tensor1, tensor2):
    """Element-wise addition followed by sum of all elements."""
    return tf.reduce_sum(tensor1 + tensor2)


# Define the function in PyTorch
def tensor_sum_pt(tensor1, tensor2):
    """Element-wise addition followed by sum of all elements."""
    return torch.sum(tensor1 + tensor2)


# Test TensorFlow with 1D, 2D, and 3D tensors
print("TensorFlow 1D Tensor Sum:")
print(tensor_sum_tf(tensor1_tf, tensor2_tf))
print("--------------------------------")

print("TensorFlow 2D Tensor Sum:")
print(tensor_sum_tf(tensor1_tf_2d, tensor2_tf_2d))
print("--------------------------------")

print("TensorFlow 3D Tensor Sum:")
print(tensor_sum_tf(tensor1_tf_3d, tensor2_tf_3d))
print("--------------------------------")

# Test PyTorch with 1D, 2D, and 3D tensors
print("PyTorch 1D Tensor Sum:")
print(tensor_sum_pt(tensor1_pt, tensor2_pt))
print("--------------------------------")

print("PyTorch 2D Tensor Sum:")
print(tensor_sum_pt(tensor1_pt_2d, tensor2_pt_2d))
print("--------------------------------")

print("PyTorch 3D Tensor Sum:")
print(tensor_sum_pt(tensor1_pt_3d, tensor2_pt_3d))


````

## Part 2 of the Assignment:

The Role of Broadcasting in Tensor Operations

Broadcasting is a powerful mechanism that allows TensorFlow and PyTorch to work with tensors of different shapes during arithmetic operations.

2.1. Describe how broadcasting works in tensor operations, providing a simple example to illustrate your explanation.

Answer: 

Broadcasting allows tensors with different shapes to work together in element-wise operations. It automatically expands the smaller tensor along dimensions of size 1 or missing dimensions so that both tensors have compatible shapes. TensorFlow and PyTorch compare the dimensions from right to left. Two dimensions are compatible if they are the same size or if one of them is 1. In most cases, broadcasting does not create a new expanded tensor in memory. Instead, the operation is performed as if the smaller tensor had been repeated.


Example:
```python
# TensorFlow
a = tf.constant([[1, 2, 3],
                 [4, 5, 6]])   # shape (2, 3)
b = tf.constant([10, 20, 30])  # shape (3,)

# b is broadcast to shape (2, 3) and added row-wise
result = a + b
# [[11, 22, 33],
#  [14, 25, 36]]
```

2.2. Discuss why broadcasting is important and how it enhances the flexibility of tensor operations in building neural network models.

Answer: 

Broadcasting is important because neural networks often work with tensors that have different shapes and sizes. For example, a batch of activations may need to be combined with a bias vector or multiplied by a scalar learning rate. Without broadcasting, we would have to manually reshape or duplicate tensors before each operation. Broadcasting makes these operations simpler by automatically matching compatible dimensions. For example, a bias with shape `(features,)` can be added directly to a batch with shape `(batch_size, features)`. This makes the code shorter, reduces the risk of shape-related errors, and improves readability while keeping the mathematical operations clear.


2.3. Explain what limitations or challenges might arise when relying on broadcasting, particularly in the context of ensuring model correctness and efficiency.

Answer: 

One challenge of broadcasting is that it can cause silent shape mismatches. If two dimensions are neither the same size nor equal to 1, the operation will fail. However, broadcasting may also expand a tensor in an unintended way, allowing the code to run while producing incorrect results. For example, adding a vector of length 4 to a `(3, 4)` matrix is valid, but adding a vector of length 3 to the same matrix is not.

Even when broadcasting works correctly, it can hide logical mistakes, such as applying a bias to the wrong axis. In terms of performance, broadcasting may appear to be an inexpensive operation, but it can still require repeated access to the smaller tensor. In some cases, using `expand_dims`, `tile`, or `repeat` makes the code easier to understand and debug. In large neural network models, relying too much on implicit broadcasting can also make performance more difficult to analyze, especially on GPU or TPU hardware where memory access patterns affect efficiency.
